2. Create Your Own Dataset
Below I constructed a 5 x 50 table (5 columns x 50 rows) using Python lists or dictionaries and random data generation packages to create my dataset. Doing so, The dataset includes different data types (e.g., integers, floats, strings, dates, and booleans) and at the end I create a DataFrame with personalized student names. 

```python
# Creating a dataset using Python 
student_records = {
    "ID_Number": [i for i in range(3001, 3051)], # unique IDs, integers
    "Exam_Score": [round(random.uniform(55,98), 1) for _ in range(50)], # scores, floats
    "Field_of_Study": [random.choice(["CS" , "Math", "Physics", "Biology"]) for _ in range(50)], # majors, strings
    "Year_of_Grad": [random.choice([2024, 2025, 2026, 2027]) for _ in range(50)], # graduation years, int
    "Passed": [random.choice([True, False]) for _ in range(50)], #pass/fail boolean
    "Enrollment_Date": [datetime.today() - timedelta(days=random.randint(0, 1200)) for _ in range(50)] # dates
}

# Create a DataFrame with student records
df_students = pd.DataFrame(student_records)
```

```python
# Display the first few rows of the dataset
df_students.head()

# Check structure and data types
df_students.info()
```

3. Data Exploration with Pandas
I will now display the first and last 5 rows of the DataFrame. In the display I will show column names and data types, generate descriptive statistics using (describe()) for numeric columns, and find the number of unique values in a categorial column. 

```python
# Preview the first and last entries in the dataset
print( "First five student records.")
display(df_students.head())

print( "\n Last five student records.")
display(df_students.tail())

# Check column details
print("\n Column details:")
print(df_students.dtypes)

# Get descriptive statistics for numeric columns
print( "\n Descriptive statistics:")
display(df_students.describe())

# Unique major count
num_unique_fields = df_students["Field_of_Study"].nunique()
print(f"\nNumber of unique fields of study: {num_unique_fields}")
```

4. Column Challenge
In this section, I will add a new column that is computed from other columns, such as converting score into a letter grade A, B, C, D, or F. 

```python
# Function to convert exam scores to grades
def convert_to_grade(score_value):
    if score_value >= 90:
        return "A"
    elif score_value >= 80:
        return "B"
    elif score_value >= 70:
        return "C"
    elif score_value >= 60:
        return "D"
    else:
        return "F"
        
# Adding a computed grade column
df_students["Letter_Grade"] = df_students["Exam_Score"].apply(convert_to_grade)

# Preview updated dataset
display(df_students.head())

# Save to CSV with student records
df_students.to_csv("student_records_lab1.csv", index=False)
```
